```python
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car=pd.read_csv('/content/quikr_car.csv')
```

```python
car.head()
```

```python
car.isnull().sum()
```

```python
car=car.dropna()
```

```python
car.isnull().sum()
```

```python
car.info()
```

```python
backup=car.copy()
```

```python
car['Price'].unique()
```

```python
# Filter out rows with "Ask For Price"
car = car[car['Price'] != 'Ask For Price']

# Reset the index if necessary
car = car.reset_index(drop=True)
```

```python
car['Price'] = car['Price'].str.replace(',', '').astype(int)
```

```python
car.info()
```

```python
car['year'].unique()
```

```python

# Extract rows where the 'year' column contains only digits
car = car[car['year'].str.isdigit()]

# Convert the 'year' column to integers
car['year'] = car['year'].astype(int)

```

```python
car.info()
```

```python
car['kms_driven'].unique()
```

```python
# Remove "km" strings and any non-digit characters
car['kms_driven'] = car['kms_driven'].str.replace('kms', '', regex=False)  # or regex=True if more 'km' patterns exist
car['kms_driven'] = car['kms_driven'].str.replace('[^0-9]', '', regex=True)

# Handle the 'petrol' string (if it's in a specific row)
car['kms_driven'] = car['kms_driven'].str.replace('petrol', '', regex=False)

# Convert the column to integers
car['kms_driven'] = pd.to_numeric(car['kms_driven'], errors='coerce').fillna(0).astype(int)

# Reset the index if necessary
car.reset_index(drop=True, inplace=True)
```

```python
car.info()
```

```python
car['fuel_type'].unique()
```

```python
car['fuel_type']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
car['fuel_type']=LabelEncoder().fit_transform(car['fuel_type'])
```

```python
car.info()
```

```python
car.isnull().sum()
```

```python
car['name'].unique()
```

```python
car['name'] = car['name'].str.split().str[1:3].str.join(' ')
```

```python
car.head()
```

```python
car
```

```python
car=car[car['Price']<6000000].reset_index(drop=True)
```

```python
x=car.drop(columns='Price')
y=car['Price']
```

```python
from sklearn.model_selection import train_test_split

```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=433)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(x[['name','company']])
```

```python
ohe.categories_
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company']),
                                    remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```
