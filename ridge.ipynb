```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
df=pd.read_csv('/content/bangalore house price prediction OHE-data 2.csv')
```

```python
df
```

deleting excess columns

```python
df=df[['bath','balcony','price','total_sqft_int','bhk','price_per_sqft','area_typeSuper built-up  Area','area_typeBuilt-up  Area','area_typePlot  Area','availability_Ready To Move']] # Use a tab character (\t) in the column name
```

```python
df.head()
```

```python
df
```

```python
x=df.drop('price',axis=1)
y=df['price']
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)

```

```python
x
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
x=poly.fit_transform(x)
```

```python
x
```

Ridge model

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import Ridge
model=Ridge()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

linear regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_predict=lr.predict(x_test)
```

```python
y_predict
```

```python
r2_score(y_test,y_predict)
```
